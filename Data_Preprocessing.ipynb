# Median Filtering

```python
from utils import open_spectra_image

image_names=['gogo', 'saintjohn', 'dionisios', 'fanourios', 'odigitria', 'minos']
file_path = f'/home/igeor/MSC-THESIS/data/h5/spec/{image_names[4]}.hdf5'

spectra_image = open_spectra_image(file_path)
spectra_image = spectra_image[100:300, 100:300, :]
```

```python
from Data_Preprocessing.utils import median_filter

filtered_image = median_filter(spectra_image, 10)
```

```python
import matplotlib.pyplot as plt 

# plot the sum of the spectra for both the original and filtered image
fig, ax = plt.subplots(2, 1, figsize=(20, 10))
ax[0].plot(spectra_image.sum(axis=(0, 1)))
ax[0].set_title('Original')
ax[1].plot(filtered_image.sum(axis=(0, 1)))
ax[1].set_title('Filtered')
```

```python
# display the original and filtered image summing bands
fig, ax = plt.subplots(1, 2, figsize=(10, 20))
ax[0].imshow(spectra_image.sum(axis=2))
ax[0].set_title('Original')
ax[1].imshow(filtered_image.sum(axis=2))
ax[1].set_title('Filtered')
```
